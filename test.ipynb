```python
import pandas as pd
import numpy as np
```

```python
import os
from dotenv import load_dotenv

load_dotenv()
DATA_PATH = os.getenv("DATA_PATH")
DATA2_PATH = os.getenv("DATA2_PATH")
```

```python
csv = DATA_PATH + "/Account.csv"
df1 = pd.read_csv(csv, delimiter=",", encoding="utf-8", keep_default_na=True, na_values=[""])
# df = df.replace({np.nan: None})
# df = df.replace({"": None})
df1
```

```python
csv = DATA2_PATH + "/Account.csv"
df2 = pd.read_csv(csv, delimiter=",", encoding="utf-8", keep_default_na=True, na_values=[""])
# df2 = df2.replace({np.nan: None})
# df2 = df2.replace({"": None})
df2
```

```python
merged_df = pd.concat([df1, df2], ignore_index=True).sort_values(df1.columns[0]).drop_duplicates(df1.columns[0], keep='last')
merged_df
```

```python
folder_name = '../DataMerged'

# Check if the folder exists, and create it if it doesn't
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
```

```python
file_path = os.path.join(folder_name, 'merged_data.csv')
merged_df.to_csv('merged_data.csv', index=False)
```

```python
df1_folder = DATA_PATH
df2_folder = DATA2_PATH
folder_name = '../DataMerged'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
for filename in os.listdir(df1_folder):
    if filename.endswith('.csv'):
        df1_file = os.path.join(df1_folder, filename)
        df2_file = os.path.join(df2_folder, filename)
        
        print(df1_file)
        if not Path(df2_file).is_file():
            # slug = '_'.join(re.findall(r'\w+', filename.lower()))
            # slug_file = os.path.join(df2_folder, slug)
            # print("checking slug", slug)
            # if not Path(slug_file).is_file():
            print(filename, "does not have a matching file in", df2_file)

            continue
        df1 = pd.read_csv(df1_file, delimiter=",", encoding="utf-8", keep_default_na=True, na_values=[""])
        df2 = pd.read_csv(df2_file, delimiter=",", encoding="utf-8", keep_default_na=True, na_values=[""])
        merged_df = pd.concat([df1, df2], ignore_index=True).sort_values(df1.columns[0]).drop_duplicates(df1.columns[0], keep='last')
        file_path = os.path.join(folder_name, filename)
        merged_df.to_csv(file_path, index=False)
```
